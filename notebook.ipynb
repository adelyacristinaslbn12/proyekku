## Visualization & Explanatory Analysis

### Pertanyaan 1:
Berapa persentase rata-rata jumlah pengguna sepeda berdasarkan cuaca?

```python
day_df['dteday'] = pd.to_datetime(day_df['dteday'])
data_by_month = day_df.groupby(day_df['dteday'].dt.to_period('M')).agg({'cnt':'sum'}).reset_index()
```

```python
result_mean_by_weathersit = day_df.groupby(by="weathersit")["cnt"].mean().reset_index()
result_mean_by_weathersit
```

```python
labels = ("clear","mist"," light rain/snow")
plt.pie(
    x=result_mean_by_weathersit["cnt"],
    labels=labels,
    autopct='%1.1f%%'
)
plt.show()
```

Pertanyaan 2.
Berapa perbandingan penggunaan sepeda pada working day dan non-working day

```python
hour_df.head(3)
```

```python
result = hour_df.groupby(["workingday"])["cnt"].sum()
result
```

```python
hari = ["Non-Working Day", "Working Day"]  # Rename the labels to reflect working day vs non-working day
plt.figure(figsize=(8, 6))
plt.bar(x=hari, height=result)
plt.title('Total Penyewaan Sepeda Berdasarkan Hari Kerja', fontsize=16)
plt.xlabel('Tipe Hari', fontsize=12)
plt.ylabel('Total Penyewaan Sepeda', fontsize=12)
plt.show()
```
